```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path6 = '/content/drive/MyDrive/Challenges Data/march-machine-learning-mania-2025/OriouVF.csv'
df6 = pd.read_csv(path6)

df6 = df6.drop(columns=['Unnamed: 0'])
```

```python
df6
```

```python
label = 'T1Won'

def evaluate_model(mod, data, features):
  model = mod
  data_train = data[data['Season'] != 2024]
  data_test = data[data['Season'] == 2024]
  X_train = data_train[features]
  y_train = data_train[label]
  X_test = data_test[features]
  y_test = data_test[label]
  model.fit(X_train, y_train)
  pred = model.predict(X_test)
  pred_proba = model.predict_proba(X_test)[:,1]
  acc = (pred == y_test).mean()
  brier = ((pred_proba - y_test)**2).mean()
  return f"accuracy : {acc} , brier score : {brier}"
```

```python
from sklearn.ensemble import RandomForestClassifier

features = ['T2_seed', 'T1_seed', 'T1_PointDiffmean', 'T2_PointDiffmean',
       'T2_Scoremean_mn3s', 'T1_FGMmean', 'T1_Scoremean_mn3s',
       'T2_FGMmean', 'T1_PFmean', 'T2_opponent_Scoremean',
       'T1_opponent_Scoremean', 'T2_Scoremean', 'T1_Scoremean',
       'T2_TOmean', 'T1_TOmean', 'T2_PFmean', 'T1_Astmean',
       'T1_opponent_ORmean', 'T2_TOQUETOQUEmean', 'T1_opponent_Stlmean',
       'T2_FGAmean_mn3s', 'T2_FGAmean', 'T1_opponent_FGMmean',
       'T1_FGAmean_mn3s', 'T2_Stlmean', 'T1_ORmean', 'T2_Astmean',
       'T2_opponent_Blkmean', 'T2_opponent_Stlmean',
       'T1_opponent_EFFGmean', 'T1_quality', 'T2_quality']

RF1 = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    bootstrap=True,
    random_state=42
)
evaluate_model(RF1,df6,features)
```

```python
path7 = '/content/drive/MyDrive/Challenges Data/march-machine-learning-mania-2025/MMasseyOrdinals.csv'
df7 = pd.read_csv(path7)
```

```python
df7
```

```python
df7 = df7[df7['RankingDayNum'].between(93, 133)]
```

```python
df7 = df7.groupby(['Season', 'TeamID', 'SystemName'])['OrdinalRank'].mean().reset_index()
```

```python
df7
```

```python
df7 = df7.groupby(['Season', 'TeamID'])['OrdinalRank'].mean().reset_index()
df7.rename(columns={'OrdinalRank': 'MeanOrdinalRank'}, inplace=True)
```

```python
df7
```

```python
# Joindre les rankings sur Team1 (T1_ID)
df6 = df6.merge(df7, left_on=['Season', 'T1_TeamID'], right_on=['Season', 'TeamID'], how='left')
df6.rename(columns={'MeanOrdinalRank': 'MeanOrdinalRank_Team1'}, inplace=True)
df6.drop(columns=['TeamID'], inplace=True)

# Joindre les rankings sur Team2 (T2_ID)
df6 = df6.merge(df7, left_on=['Season', 'T2_TeamID'], right_on=['Season', 'TeamID'], how='left')
df6.rename(columns={'MeanOrdinalRank': 'MeanOrdinalRank_Team2'}, inplace=True)
df6.drop(columns=['TeamID'], inplace=True)

# Calculer la différence de ranking
df6['OrdinalRank_Diff'] = df6['MeanOrdinalRank_Team1'] - df6['MeanOrdinalRank_Team2']
```
