# Feature Coach_changed

```python
# On compte le nombre de lignes par équipe dans une saison (c'est enfait le nombre de coachs)
coach_changes = df_coaches.groupby(['Season', 'TeamID']).size().reset_index(name='Coach_Changed')
```

```python
# Si une équipe a eu plus d'un coach dans la saison, mettre 1 sinon 0
coach_changes['Coach_Changed'] = (coach_changes['Coach_Changed'] > 1).astype(int)
```

```python
df6 = df6.merge(coach_changes, left_on=['Season', 'T1_TeamID'], right_on=['Season', 'TeamID'], how='left')
df6.rename(columns={'Coach_Changed' : 'Coach_Changed_Team1'}, inplace=True)
df6.drop(columns=['TeamID'], inplace=True)
```

```python
df6 = df6.merge(coach_changes, left_on=['Season', 'T2_TeamID'], right_on=['Season', 'TeamID'], how='left')
df6.rename(columns={'Coach_Changed' : 'Coach_Changed_Team2'}, inplace=True)
df6.drop(columns=['TeamID'], inplace=True)
```

```python
df6
```

```python
#Est ce qu'on garde les coach change pour les deux Team ou est ce qu'on garde que la diff ? ou les deux (solution hybride)
df6['Coach_Changed_Diff'] = df6['Coach_Changed_Team1'] - df6['Coach_Changed_Team2']
```

```python
df6
```

# Evaluate model

```python
label = 'T1Won'

def evaluate_model(mod, data, features):
  model = mod
  data_train = data[data['Season'] != 2024]
  data_test = data[data['Season'] == 2024]
  X_train = data_train[features]
  y_train = data_train[label]
  X_test = data_test[features]
  y_test = data_test[label]
  model.fit(X_train, y_train)
  pred = model.predict(X_test)
  pred_proba = model.predict_proba(X_test)[:,1]
  acc = (pred == y_test).mean()
  brier = ((pred_proba - y_test)**2).mean()
  return f"accuracy : {acc} , brier score : {brier}"
```

```python
from sklearn.ensemble import RandomForestClassifier

features = ['T2_seed', 'T1_seed', 'T1_PointDiffmean', 'T2_PointDiffmean',
       'T2_Scoremean_mn3s', 'T1_FGMmean', 'T1_Scoremean_mn3s',
       'T2_FGMmean', 'T1_PFmean', 'T2_opponent_Scoremean',
       'T1_opponent_Scoremean', 'T2_Scoremean', 'T1_Scoremean',
       'T2_TOmean', 'T1_TOmean', 'T2_PFmean', 'T1_Astmean',
       'T1_opponent_ORmean', 'T2_TOQUETOQUEmean', 'T1_opponent_Stlmean',
       'T2_FGAmean_mn3s', 'T2_FGAmean', 'T1_opponent_FGMmean',
       'T1_FGAmean_mn3s', 'T2_Stlmean', 'T1_ORmean', 'T2_Astmean',
       'T2_opponent_Blkmean', 'T2_opponent_Stlmean',
       'T1_opponent_EFFGmean', 'T1_quality', 'T2_quality', 'Coach_Changed_Team1', 'Coach_Changed_Team2', 'Coach_Changed_Diff']

RF1 = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    bootstrap=True,
    random_state=42
)
evaluate_model(RF1,df6,features)
```

```python
df6.drop(columns=['Coach_Changed_Team1', 'Coach_Changed_Team2', 'Coach_Changed_Diff'], inplace=True)
```
